```python
#03. Scaling - min max normalization
```

```python
(patient_df['height'] - patient_df['height'].min()) / \
(patient_df['height'].max() - patient_df['height'].min())
```

```python
patient_df['height'] = (patient_df['height'] - patient_df['height'].min()) / \
                       (patient_df['height'].max() - patient_df['height'].min())
```
