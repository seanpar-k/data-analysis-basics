```python
# Scaling - min max normalization
```

```python
(patient_df['height'] - patient_df['height'].min()) / \
(patient_df['height'].max() - patient_df['height'].min())
```

```python
patient_df['height'] = (patient_df['height'] - patient_df['height'].min()) / \
                       (patient_df['height'].max() - patient_df['height'].min())
```

```python
patient_df
```

```python
patient_df['height'].describe()
```

```python
# Scaling - standardization
```

```python
# Z-score
```

```python
(patient_df['height'] - patient_df['height'].mean()) / patient_df['height'].std()
```

```python
patient_df['height'] = (patient_df['height'] - patient_df['height'].mean()) / \
                       patient_df['height'].std()
```

```python
def group_age(x):
    if x>= 10 and x < 20:
        return '10s'
    elif x >=20 and x < 30:
        return '20s'
    elif x >= 30 and x < 40:
        return '30s'
    elif x >= 40 and x < 50:
        return '40s'
    elif x >= 50 and x < 60:
        return '50s'
    else:
        return '60s'
```

```python
patient_df['age_group'] = patient_df['age'].apply(group_age)
```

```python
patient_df
```

```python
patient_df['bmi'] = round(patient_df['weight'] / patient_df['height']**2, 1)
patient_df['bmi'].apply(lambda x: 'Y' if x > 30 else 'N')
```

```python
def isin_interval(x, n_lower, n_upper):
    if n_lower <= x < n_upper:
        return 'Y'
    else:
        return 'N'
```

```python
patient_df['bmi'].apply(lambda x: isin_interval(x, 18.5, 25))
```
