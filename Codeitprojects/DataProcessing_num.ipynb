```python
# Scaling - min max normalization
```

```python
(patient_df['height'] - patient_df['height'].min()) / \
(patient_df['height'].max() - patient_df['height'].min())
```

```python
patient_df['height'] = (patient_df['height'] - patient_df['height'].min()) / \
                       (patient_df['height'].max() - patient_df['height'].min())
```

```python
patient_df
```

```python
patient_df['height'].describe()
```

```python
# Scaling - standardization
```

```python
# Z-score
```

```python
(patient_df['height'] - patient_df['height'].mean()) / patient_df['height'].std()
```

```python
patient_df['height'] = (patient_df['height'] - patient_df['height'].mean()) / \
                       patient_df['height'].std()
```

```python
patient_df
```

```python
patient_df['height'].describe()
```

```python
# normalization   -> 데이터의 범위를 0과 1 사이 값으로 통일
# standardization -> 각 데이터가 평균에서 얼만큼 떨어져 있는지 비교
```
